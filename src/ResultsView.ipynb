```python

import pandas as pd
import numpy as np
from matplotlib import gridspec
import math
```

```python
df = pd.read_csv("CSVResult/CSVResultFinal.csv")
df = df.drop(0).reset_index()
df = df.drop(columns = ['degree', 'coef',
 'Validation foldN _0',
 'Validation foldN _1',
 'Validation foldN _2',
 'Validation foldN _3',
 'Validation foldN _4',
 'Training foldN _0',
 'Training foldN _1',
 'Training foldN _2',
 'Training foldN _3',
 'Training foldN _4', 'time','kernel'])
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
df=df.astype('float')
```

```python
C=np.log10(df.values[:,5]);
gamma=np.log10(df.values[:,6]);
epsilon=np.log10(df.values[:,7]);
validationresult=df.values[:,1];
trainingresult=df.values[:,3];
trainingresult;
```

```python
fig = plt.figure(figsize=(11,8))
ax = fig.add_subplot(122,projection='3d')
ax.set_ylabel("log10(gamma)")
ax.set_zlabel("log10(epsilon)")
ax.set_xlabel("log10(C)")



surf=ax.scatter(C, gamma, epsilon, c=validationresult, cmap=plt.cool())
fig.colorbar(surf, shrink=0.5, aspect=5)


ax2 = fig.add_subplot(121,projection='3d')
ax2.set_ylabel("log10(gamma)")
ax2.set_zlabel("log10(epsilon)")
ax2.set_xlabel("log10(C)")
surf2=ax.scatter(C, gamma, epsilon, c=trainingresult, cmap=plt.cool())
fig.colorbar(surf2, shrink=1, aspect=1)


plt.show()
```
