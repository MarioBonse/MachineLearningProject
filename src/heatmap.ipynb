```python
df = pd.read_csv("CSVResult/CSVResultFinal.csv")
df = df.drop(0).reset_index()
df = df.drop(columns = ['degree', 'coef',
  'time', 'epsilon', 'index',
     'kernel'])
df = df.astype('float')
df = df.round(3)
```

```python
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
fmri = sns.load_dataset("fmri")
ax = sns.lineplot(x="C", y="Validation mean", data=df)
```

```python
df_04 = df.loc[df['gamma'] == 0.4]
df_05 = df.loc[df['gamma'] == 0.5]
df_06 = df.loc[df['gamma'] == 0.6]
```

```python
df_04_0 = df_04[["C", "Validation foldN _0", 'Training foldN _0']].rename(index=str, columns={"Validation foldN _0":"validationerror",'Training foldN _0':"trainingerror"})
df_04_1 = df_04[["C", "Validation foldN _1", 'Training foldN _1']].rename(index=str, columns={"Validation foldN _1":"validationerror",'Training foldN _1':"trainingerror"})
df_04_2 = df_04[["C", "Validation foldN _2", 'Training foldN _2']].rename(index=str, columns={"Validation foldN _2":"validationerror",'Training foldN _2':"trainingerror"})
df_04_3 = df_04[["C", "Validation foldN _3", 'Training foldN _3']].rename(index=str, columns={"Validation foldN _3":"validationerror",'Training foldN _3':"trainingerror"})
df_04_4 = df_04[["C", "Validation foldN _4", 'Training foldN _4']].rename(index=str, columns={"Validation foldN _4":"validationerror",'Training foldN _4':"trainingerror"})
result04 = pd.concat([df_04_0,df_04_1,df_04_2,df_04_3,df_04_4]).reset_index()
```

```python
df_04_0 = df_05[["C", "Validation foldN _0", 'Training foldN _0']].rename(index=str, columns={"Validation foldN _0":"validationerror",'Training foldN _0':"trainingerror"})
df_04_1 = df_05[["C", "Validation foldN _1", 'Training foldN _1']].rename(index=str, columns={"Validation foldN _1":"validationerror",'Training foldN _1':"trainingerror"})
df_04_2 = df_05[["C", "Validation foldN _2", 'Training foldN _2']].rename(index=str, columns={"Validation foldN _2":"validationerror",'Training foldN _2':"trainingerror"})
df_04_3 = df_05[["C", "Validation foldN _3", 'Training foldN _3']].rename(index=str, columns={"Validation foldN _3":"validationerror",'Training foldN _3':"trainingerror"})
df_04_4 = df_05[["C", "Validation foldN _4", 'Training foldN _4']].rename(index=str, columns={"Validation foldN _4":"validationerror",'Training foldN _4':"trainingerror"})
result05 = pd.concat([df_04_0,df_04_1,df_04_2,df_04_3,df_04_4]).reset_index()
```

```python
df_04_0 = df_06[["C", "Validation foldN _0", 'Training foldN _0']].rename(index=str, columns={"Validation foldN _0":"validationerror",'Training foldN _0':"trainingerror"})
df_04_1 = df_06[["C", "Validation foldN _1", 'Training foldN _1']].rename(index=str, columns={"Validation foldN _1":"validationerror",'Training foldN _1':"trainingerror"})
df_04_2 = df_06[["C", "Validation foldN _2", 'Training foldN _2']].rename(index=str, columns={"Validation foldN _2":"validationerror",'Training foldN _2':"trainingerror"})
df_04_3 = df_06[["C", "Validation foldN _3", 'Training foldN _3']].rename(index=str, columns={"Validation foldN _3":"validationerror",'Training foldN _3':"trainingerror"})
df_04_4 = df_06[["C", "Validation foldN _4", 'Training foldN _4']].rename(index=str, columns={"Validation foldN _4":"validationerror",'Training foldN _4':"trainingerror"})
result06 = pd.concat([df_04_0,df_04_1,df_04_2,df_04_3,df_04_4]).reset_index()
```

```python
ax = sns.lineplot(x="C", y="validationerror",data=result04)
ax = sns.lineplot(x="C", y="trainingerror",data=result04)
```

```python
ax = sns.lineplot(x="C", y="validationerror",data=result05)
ax = sns.lineplot(x="C", y="trainingerror",data=result05, dashes=True)
#ax = sns.lineplot(x="C", y="trainingerror", dashes=True, kind="line", data=result05);
ax.legend(["validationerror", "trainingerror"])
```

```python
ax = sns.lineplot(x="C", y="validationerror",data=result06)
ax = sns.lineplot(x="C", y="trainingerror",data=result06)
ax.legend(["validationerror", "trainingerror"])
```
