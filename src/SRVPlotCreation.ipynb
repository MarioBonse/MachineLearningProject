```python
import pandas as pd
import numpy as np
from matplotlib import gridspec
import math
# This file jupyter uses the results obtained with the grid search in order to plot the 
# results we have had with svm
```

```python
df = pd.read_csv("CSVResult/CSVResultFinal.csv")
df = df.drop(0).reset_index()
df = df.drop(columns = ['degree', 'coef',
 'Validation foldN _0',
 'Validation foldN _1',
 'Validation foldN _2',
 'Validation foldN _3',
 'Validation foldN _4',
 'Training foldN _0',
 'Training foldN _1',
 'Training foldN _2',
 'Training foldN _3',
 'Training foldN _4', 'time','kernel'])
df=df.astype('float')
```

```python
import matplotlib.pyplot as plt
df4 = df.loc[df["gamma"] == 0.4]
plt.plot(df4["C"], df4["Validation mean"], '-b', label='valiadtion error(mean)')

plt.plot(df4["C"], df4["Validation deviation"], '-.y', label='valiadtion error(std)')
plt.plot(df4["C"], df4["Training mean"], '--r', label='training error')
plt.xlabel("Alpha")
plt.ylabel("MEE")
plt.title("Gamma = 0.4")
plt.legend(loc = 'upper right')
plt.show()
```

```python
import matplotlib.pyplot as plt
df5 = df.loc[df["gamma"] == 0.5]
plt.plot(df5["C"], df5["Validation mean"], '-b', label='valiadtion error(mean)')
plt.plot(df5["C"], df5["Validation deviation"], '-.y', label='valiadtion error(std)')
plt.plot(df5["C"], df5["Training mean"], '--r', label='training error')

plt.xlabel("Alpha")
plt.ylabel("MEE")
plt.title("Gamma = 0.5")
plt.legend(loc = 'upper right')
plt.show()
```
