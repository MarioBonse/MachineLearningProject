```python
import pandas as pd
import numpy as np
from matplotlib import gridspec
import math
import seaborn as sns
# This file jupyter uses the results obtained with the grid search in order to plot the 
# results we have had with svm
```

```python
df = pd.read_csv("../result/RBF_C_Gammafixed.csv")
df = df.drop(columns = ['degree', 'coef',
 'Validation foldN _0',
 'Validation foldN _1',
 'Validation foldN _2',
 'Validation foldN _3',
 'Validation foldN _4',
 'Training foldN _0',
 'Training foldN _1',
 'Training foldN _2',
 'Training foldN _3',
 'Training foldN _4', 'time','kernel'])
df=df.astype('float')
```

```python
df.sort_values("Validation mean")
```

```python
import matplotlib.pyplot as plt
Gamma006 = df.loc[df["gamma"] == 0.06].reset_index(drop = True)
Gamma006 = Gamma006.astype('float')
Gamma006 = Gamma006.sort_values(by = ("C"))
plt.plot(Gamma006["C"], Gamma006["Validation mean"], '-b', label='valiadtion error(mean)')
plt.plot(Gamma006["C"], Gamma006["Validation deviation"], '-.y', label='valiadtion error(std)')
plt.plot(Gamma006["C"], Gamma006["Training mean"], '--r', label='training error')
plt.xlabel("C")
plt.ylabel("MEE")
plt.title("Gamma = 0.06")
plt.legend(loc = 'upper right')
plt.show()
```

```python
import matplotlib.pyplot as plt
Gamma01 = df.loc[df["gamma"] == 0.1]
Gamma01 = Gamma01.astype('float')
Gamma01 = Gamma01.sort_values(by = ("C"))
plt.plot(Gamma01["C"], Gamma01["Validation mean"], '-b', label='valiadtion error(mean)')
plt.plot(Gamma01["C"], Gamma01["Validation deviation"], '-.y', label='valiadtion error(std)')
plt.plot(Gamma01["C"], Gamma01["Training mean"], '--r', label='training error')

plt.xlabel("C")
plt.ylabel("MEE")
plt.title("Gamma = 0.1")
plt.legend(loc = 'upper right')
plt.show()
```
